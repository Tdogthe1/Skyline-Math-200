```python
df['Military_Time'] = df['H4SP1H']
df['Military_Time']= pd.to_numeric(df['Military_Time'],errors= 'coerce')
df['Military_Time']= df['Military_Time'].loc[df['Military_Time'] < 96]
df['Military_Time']= df['Military_Time'].loc[df['Military_Time'] != ' ']
df['Military_Time'].value_counts()
```

```python
df_time_convert = df[['am_pm', 'Military_Time', 'Wake_Up_minute']]
```

```python
def to_military_time_float(in_hour,in_min,am_pm):
  out_time = 0
  if in_hour == '12':

    if am_pm == 'AM': 
      out_time = 0
    else:
      out_time = int(in_hour)
  
  elif (0 < int(in_hour) < 12):
    if am_pm == 'AM':
      out_time = int(in_hour)
    else:
      out_time = int(in_hour) + 12
  
  

  out_min = int(in_min) / 60

  out_time += out_min 




  return out_time
```
